```python
from Bio.SeqUtils import ProtParam
from Bio.Seq import Seq
#If you want, add the parameters more 

protein_sequence = Seq("TFIDVDCTVSKECWAPCKAAFGVDRGKCMGKKCKCY") 
lowercase = str(protein_sequence).lower()
prot_param = ProtParam.ProteinAnalysis(lowercase)
molecular_weight = prot_param.molecular_weight()
isoelectric_point = prot_param.isoelectric_point()
gravy_value = prot_param.gravy()
aminoacidcontent = prot_param.count_amino_acids()
aminoacidspercent = prot_param.get_amino_acids_percent()
protein_charge = prot_param.charge_at_pH(7.4)
aminoacids_count = prot_param.count_amino_acids()
protein_instability_index = prot_param.instability_index()
protein_aromaticity = prot_param.aromaticity()
stability_index = prot_param.instability_index()
pI = prot_param.isoelectric_point()
asp_count = prot_param.count_amino_acids()['D']
glu_count = prot_param.count_amino_acids()['E']
negatively_charged_residues = asp_count + glu_count

lysine_count = prot_param.count_amino_acids()['L']
arginine_count = prot_param.count_amino_acids()['R']
positively_charged_residues = lysine_count + arginine_count


isoleucine_percent = prot_param.get_amino_acids_percent()['I']/100
valine_percent = prot_param.get_amino_acids_percent()['V']/100
leucine_percent = prot_param.get_amino_acids_percent()['L']/100


print(f"Molecular weight: {molecular_weight:.2f}")
print("-------------------------")

print(f"Isoelectric point: {isoelectric_point:.2f}")
print("-------------------------")

print(f"gravy value: {gravy_value:.2f}")
print("-------------------------")

print(f"Charge at pH 7.4: {protein_charge:.2f}")

print("-------------------------")

print(f"Instability index: {protein_instability_index:.2f}")

print("-------------------------")

print(f"Aromaticity of protein: {protein_aromaticity:.2f}")

print("-------------------------")

print(f"Theoretical pI: {pI:.2f}")

print("-------------------------")    

print(f"Total number of negatively charged residues (Asp + Glu): {negatively_charged_residues}")

print("-------------------------")   

print(f"Total number of positively charged residues (Arg + Lys): {positively_charged_residues}")

print("-------------------------")    

for amino_acid, percent in aminoacidspercent.items():
    print(f"{amino_acid}: {percent:.2f}%")
print("-------------------------")

for amino_acid, count in aminoacidcontent.items():
    print(f"{amino_acid}: {count}")
```
